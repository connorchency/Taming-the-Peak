```python
data1 = pd.DataFrame()
# getting Dec 2013 data
for i in range (1,32):
    nrg = pd.read_csv('https://www.wunderground.com/history/airport/EGLL/2013/12/' + str(i) + '/DailyHistory.html?req_city=London&req_state=GLA&req_statename=United%20Kingdom&reqdb.zip=00000&reqdb.magic=1&reqdb.wmo=03772&MR=1&format=1')
    nrg['Date'] = '2013/12/' + str(i)
    data1 = pd.concat([data1, nrg])

    
data1.describe()
```

```python
# getting Jan 2014 data
for i in range (1,32):
    nrg = pd.read_csv('https://www.wunderground.com/history/airport/EGLL/2014/1/' + str(i) + '/DailyHistory.html?req_city=London&req_state=GLA&req_statename=United%20Kingdom&reqdb.zip=00000&reqdb.magic=1&reqdb.wmo=03772&MR=1&format=1')
    nrg['Date'] = '2014/1/' + str(i)
    data1 = pd.concat([data1, nrg])
# getting Feb 2014 data
for i in range (1,29):
    nrg = pd.read_csv('https://www.wunderground.com/history/airport/EGLL/2014/2/' + str(i) + '/DailyHistory.html?req_city=London&req_state=GLA&req_statename=United%20Kingdom&reqdb.zip=00000&reqdb.magic=1&reqdb.wmo=03772&MR=1&format=1')
    nrg['Date'] = '2014/2/' + str(i)
    data1 = pd.concat([data1, nrg])
# getting Mar 2014 data
for i in range (1,32):
    nrg = pd.read_csv('https://www.wunderground.com/history/airport/EGLL/2014/3/' + str(i) + '/DailyHistory.html?req_city=London&req_state=GLA&req_statename=United%20Kingdom&reqdb.zip=00000&reqdb.magic=1&reqdb.wmo=03772&MR=1&format=1')
    nrg['Date'] = '2014/3/' + str(i)
    data1 = pd.concat([data1, nrg])
```
