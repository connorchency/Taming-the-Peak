```python
iop = pd.read_csv('London_weather_2011-14.csv')
iop
```

```python
weather_clean = iop[['DateUTC<br />', 'Dew PointF', 'Humidity', 'TemperatureF', 'Wind SpeedMPH']]
weather_clean.columns = [['Datetime', 'Dew_Point_F', 'Humidity', 'Temperature_F', 'Wind_Speed_MPH']]
```

```python
weather_clean
```

```python
weather_clean_r = pd.DataFrame()
for i in range(len(weather_clean)):
    if weather_clean['Datetime'][i][14:] == '50:00<br />':
        weather_clean_r = weather_clean_r.append(weather_clean.iloc[(i)][:19])
weather_clean_r.describe()
```

```python
weather_clean_r
```

```python
weather_clean_r.sort(columns='Datetime', axis=0, ascending=True, inplace=True)
weather_clean_r
```

```python
weather_clean_r['Datetime'].drop_duplicates(inplace=True)
weather_clean_r
```

```python
weather_clean_r.reset_index(inplace=True)
```

```python
weather_clean_r
```

```python
d = []
for i in range(len(weather_clean_r)-1):
    if weather_clean_r['Datetime'][i] == weather_clean_r['Datetime'][i+1]:
        d.append(i)
    
d
```

```python
len(d)
```

```python
weather_clean_r.drop('index', 1, inplace=True)
weather_clean_r
```

```python
weather_clean_r.drop(weather_clean_r.index(d), inplace=True)
weather_clean_r
```

```python
#weather_clean_r.drop(weather_clean_r.index[d], inplace=True)
weather_clean_r.reset_index(inplace=True)
weather_clean_r
```
